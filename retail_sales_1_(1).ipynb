```python
%%writefile retail_sales_app.py
import pandas as pd
import numpy as np
import streamlit as st
import plotly.express as px

# Page configuration
st.set_page_config(page_title="Retail Sales Analysis", layout="wide", page_icon="ðŸ“Š")

# Add a logo or header image (optional)
st.image('https://cdn-icons-png.flaticon.com/512/3405/3405793.png', width=200)

# Add centered title
st.markdown("<h1 style='text-align: center; color: white;'>Retail Sales Analysis Dashboard</h1>", unsafe_allow_html=True)

# Sidebar file uploader
st.sidebar.header("Upload Your Retail Sales CSV")
uploaded_file = st.sidebar.file_uploader("Choose a CSV file", type=["csv"])

if uploaded_file is not None:
    # Load data
    df = pd.read_csv(uploaded_file)
    st.dataframe(df.head())

    # Sidebar page selection
    page = st.sidebar.radio("Pages", ["Univariate Analysis", "Bivariate Analysis", "Multivariate Analysis"])

    # UNIVARIATE ANALYSIS
    if page == "Univariate Analysis":
        st.subheader("Univariate Analysis")

        col = st.selectbox("Select Column", df.columns)

        chart = st.selectbox("Select Chart", ["Histogram", "Box", "Pie"])

        if chart == "Histogram":
            st.plotly_chart(px.histogram(df, x=col, title=f"Histogram of {col}"))
        elif chart == "Box":
            st.plotly_chart(px.box(df, y=col, title=f"Box Plot of {col}"))
        elif chart == "Pie":
            if df[col].dtype == 'object' or len(df[col].unique()) < 20:
                st.plotly_chart(px.pie(df, names=col, title=f"Pie Chart of {col}"))
            else:
                st.warning("Pie charts work best with categorical data.")

    # BIVARIATE ANALYSIS
    elif page == "Bivariate Analysis":
        st.subheader("Bivariate Analysis")

        # 1ï¸âƒ£ Is there a correlation between item price and quantity sold?
        if 'Price Per Unit' in df.columns and 'Quantity' in df.columns:
            st.markdown("### ðŸ“ˆ Correlation between Item Price and Quantity Sold")
            fig1 = px.scatter(df, x='Price Per Unit', y='Quantity', title="Price vs. Quantity Sold")
            st.plotly_chart(fig1)
        else:
            st.info("Columns 'Price Per Unit' and/or 'Quantity' not found.")

        # 2ï¸âƒ£ Do certain payment methods affect larger transactions?
        if 'Payment Method' in df.columns and 'Total Spent' in df.columns:
            st.markdown("### ðŸ’³ Payment Method vs. Total Spent")
            fig2 = px.box(df, x='Payment Method', y='Total Spent', title="Payment Method vs. Total Spent")
            st.plotly_chart(fig2)
        else:
            st.info("Columns 'Payment Method' and/or 'Total Spent' not found.")

        # 3ï¸âƒ£ Are weekends more profitable than weekdays?
        if 'Weekend' in df.columns and 'Total Spent' in df.columns:
            st.markdown("### ðŸ“… Weekend vs. Total Spent")
            fig3 = px.scatter(df, x='Weekend', y='Total Spent', title="Weekend vs. Total Spent")
            st.plotly_chart(fig3)
        else:
            st.info("Columns 'Weekend' and/or 'Total Spent' not found.")

        # 4ï¸âƒ£ Which categories have the highest average transaction value?
        if 'Category' in df.columns and 'Total Spent' in df.columns:
            st.markdown("### ðŸ›’ Category vs. Total Spent")
            fig4 = px.box(df, x='Category', y='Total Spent', title="Category vs. Total Spent")
            st.plotly_chart(fig4)
        else:
            st.info("Columns 'Category' and/or 'Total Spent' not found.")

    # MULTIVARIATE ANALYSIS
    elif page == "Multivariate Analysis":
        st.subheader("Multivariate Analysis")

        numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
        if len(numeric_cols) >= 2:
            x_col = st.selectbox("X-axis", numeric_cols, index=0)
            y_col = st.selectbox("Y-axis", numeric_cols, index=1)
            color_col = st.selectbox("Color By (optional)", [None] + df.columns.tolist())
            if color_col == "None":
                color_col = None
            fig = px.scatter(df, x=x_col, y=y_col, color=color_col, title=f"{y_col} vs {x_col}")
            st.plotly_chart(fig)
        else:
            st.warning("Not enough numeric columns for multivariate analysis.")
else:
    st.info("Please upload a CSV file to get started.")

st.sidebar.markdown("---")
st.sidebar.caption("Developed with â¤ï¸ using Streamlit and Plotly.")

```
